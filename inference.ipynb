# Library Import

```python
import pandas as pd
import pickle
import random
import warnings
warnings.filterwarnings('ignore')
```

# Import Model

```python
with open('best_model.pkl','rb') as file_1:
    model = pickle.load(file_1)
```

# Make Dummy Dataframe For Testing

```python
def generate_categorical_data(options, size=20):
    return random.choices(options, k=size)

def generate_numerical_data(min_value, max_value, size=20):
    return [random.randint(min_value, max_value) for _ in range(size)]

# Data options for categorical attributes
sex_options = ['Male', 'Female']
high_school_type_options = ['State', 'Private']
scholarship_options = ['25%', '50%', '75%', '100%']
yes_no_options = ['Yes', 'No']
transportation_options = ['Bus', 'Private']
attendance_options = ['Always', 'Sometimes']
reading_options = ['Yes', 'No']
listening_options = ['Yes', 'No']
project_work_options = ['Yes', 'No']

# Generate new set of data
made_up_data = {
    'Student_Age': generate_numerical_data(18, 25),
    'Sex': generate_categorical_data(sex_options),
    'High_School_Type': generate_categorical_data(high_school_type_options),
    'Scholarship': generate_categorical_data(scholarship_options),
    'Additional_Work': generate_categorical_data(yes_no_options),
    'Sports_activity': generate_categorical_data(yes_no_options),
    'Transportation': generate_categorical_data(transportation_options),
    'Weekly_Study_Hours': generate_numerical_data(1, 20),
    'Attendance': generate_categorical_data(attendance_options),
    'Reading': generate_categorical_data(reading_options),
    'Notes': generate_categorical_data(yes_no_options),
    'Listening_in_Class': generate_categorical_data(listening_options),
    'Project_work': generate_categorical_data(project_work_options)
}
```

```python
data_inf = pd.DataFrame(made_up_data)

data_inf.info()
data_inf
```

# Predict and Save it For Analysis at main notebook

```python
# Data Predict
data_final_pred = model.predict(data_inf)
data_final_pred
```

```python
data_inf['Final Score'] = data_final_pred
data_inf
```

```python
csv_file_path = 'dummy_test.csv'
data_inf.to_csv(csv_file_path, index=False)
```
